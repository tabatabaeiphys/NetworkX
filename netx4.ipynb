## Degree Centrality

Degree Centrality measures the number of direct connections a node has.

For an undirected graph with $N$ nodes, the degree centrality of node $i$ is:

$$
C_D(i) = \frac{\deg(i)}{N - 1}
$$

Where:

- $\deg(i)$ is the **degree** (number of neighbors) of node $i$  
- $N$ is the **total number of nodes** in the graph

This value is normalized between 0 and 1.


```python
deg = DC(G)  

top_indices = np.argsort(deg)[-5:][::-1]
for i in top_indices:
    print(f"Node {i + 1}: Degree Centrality = {deg[i]:.4f}")

pos = nx.spring_layout(G, seed=42)
node_sizes = 300 + 1000 * (deg - np.min(deg)) / (np.ptp(deg) + 1e-6)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray',
        node_size=node_sizes, font_size=10, font_weight='bold')

nx.draw_networkx_nodes(G, pos,
                       nodelist=top_indices.tolist(),
                       node_color='lightgreen',
                       node_size=1500,
                       label='Top 5')

plt.title("Degree Centrality (Top 5 Highlighted)")
plt.legend()
plt.show()
```

## Influence Centrality

Influence Centrality compares a node’s degree with those of its neighbors to measure relative dominance.

Defined for node $i $ as:

$$
C_I(i) = \frac{1}{d_i} \sum_{j \in N(i)} \frac{d_i - d_j}{d_i + d_j}
$$

Where:
- $d_i$ is the degree of node $ i $,
- $N(i)$ is the set of neighbors of node $ i $.

A node has high influence if its degree is significantly larger than that of its neighbors.


```python
inf = influence_centrality(G)  

top_indices = np.argsort(inf)[-5:][::-1]
for i in top_indices:
    print(f"Node {i + 1}: Influence Centrality = {inf[i]:.4f}")

pos = nx.spring_layout(G, seed=42)
node_sizes = 300 + 1000 * (inf - np.min(inf)) / (np.ptp(inf) + 1e-6)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray',
        node_size=node_sizes, font_size=10, font_weight='bold')

nx.draw_networkx_nodes(G, pos,
                       nodelist=top_indices.tolist(),
                       node_color='deepskyblue',
                       node_size=1500,
                       label='Top 5')

plt.title("Influence Centrality (Top 5 Highlighted)")
plt.legend()
plt.show()
```

## Closeness Centrality

Closeness Centrality measures how close a node is to all other nodes in the network.

It is defined as:

$$
C_C(i) = \frac{N - 1}{\sum_{j \neq i} d(i, j)}
$$

Where:

- $d(i, j)$ is the **shortest path distance** between nodes $i$ and $j$  
- $N$ is the **total number of nodes** in the graph

Nodes with high closeness can quickly interact with all others.


```python
clo = CC(G) 

top_indices = np.argsort(clo)[-5:][::-1]
for i in top_indices:
    print(f"Node {i + 1}: Closeness Centrality = {clo[i]:.4f}")

pos = nx.spring_layout(G, seed=42)
node_sizes = 300 + 1000 * (clo - np.min(clo)) / (np.ptp(clo) + 1e-6)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray',
        node_size=node_sizes, font_size=10, font_weight='bold')

nx.draw_networkx_nodes(G, pos,
                       nodelist=top_indices.tolist(),
                       node_color='lightblue',
                       node_size=1500,
                       label='Top 5')

plt.title("Closeness Centrality (Top 5 Highlighted)")
plt.legend()
plt.show()
```

## Betweenness Centrality

Betweenness Centrality quantifies the number of times a node acts as a bridge along the shortest path between two other nodes.

The formula is:

$$
C_B(i) = \sum_{s \neq i \neq t} \frac{\sigma_{st}(i)}{\sigma_{st}}
$$

Where:

- $\sigma_{st}$ is the **total number of shortest paths** from node $s$ to node $t$  
- $\sigma_{st}(i)$ is the **number of those paths that pass through** node $i$

Nodes with high betweenness control information flow.


```python
bet = BC(G)  

top_indices = np.argsort(bet)[-5:][::-1]
for i in top_indices:
    print(f"Node {i + 1}: Betweenness Centrality = {bet[i]:.4f}")

pos = nx.spring_layout(G, seed=42)
node_sizes = 300 + 1000 * (bet - np.min(bet)) / (np.ptp(bet) + 1e-6)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray',
        node_size=node_sizes, font_size=10, font_weight='bold')

nx.draw_networkx_nodes(G, pos,
                       nodelist=top_indices.tolist(),
                       node_color='salmon',
                       node_size=1500,
                       label='Top 5')

plt.title("Betweenness Centrality (Top 5 Highlighted)")
plt.legend()
plt.show()
```

## Bridge Centrality

Bridge Centrality combines Betweenness Centrality with how structurally unique a node’s neighborhood is.

Let $B(i)$ be the betweenness centrality of node $i$, then:

$$
\text{Bridge Centrality}(i) = B(i) \cdot \left( \frac{1 / d_i}{\sum_{j \in N(i)} 1 / d_j} \right)
$$

Where:
- $d_i$ is the degree of node $i$,
- $N(i)$ is the set of neighbors of node $i$.

Nodes that act as key links between otherwise distant parts of the network tend to have high Bridge Centrality.


```python
bridge = bridge_centrality(G) 

top_indices = np.argsort(bridge)[-5:][::-1]
for i in top_indices:
    print(f"Node {i + 1}: Bridge Centrality = {bridge[i]:.4f}")

pos = nx.spring_layout(G, seed=42)
node_sizes = 300 + 1000 * (bridge - np.min(bridge)) / (np.ptp(bridge) + 1e-6)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray',
        node_size=node_sizes, font_size=10, font_weight='bold')

nx.draw_networkx_nodes(G, pos,
                       nodelist=top_indices.tolist(),
                       node_color='gold',
                       node_size=1500,
                       label='Top 5')

plt.title("Bridge Centrality (Top 5 Highlighted)")
plt.legend()
plt.show()
```

## Eigenvector Centrality

Eigenvector Centrality gives relative scores to nodes based on the idea that connections to high-scoring nodes contribute more to a node’s importance.

It satisfies the equation:

$$
\mathbf{x} = \lambda^{-1} A \mathbf{x}
$$

Where:

- $A$ is the **adjacency matrix** of the graph  
- $\lambda$ is the **largest eigenvalue** of the matrix $A$  
- $\mathbf{x}$ is the **eigenvector** corresponding to $\lambda$; its $i$‑th entry $x_i$ represents the centrality of node $i$

In other words, a node is important if it is connected to other important nodes.  
This is a **recursive definition** of centrality, and it leads to the eigenvector equation:

$$
A \mathbf{x} = \lambda \mathbf{x}
$$


```python
eig = EC(G)  
top_indices = np.argsort(eig)[-5:][::-1]
for i in top_indices:
    print(f"Node {i + 1}: Eigenvector Centrality = {eig[i]:.4f}")

pos = nx.spring_layout(G, seed=42)
node_sizes = 300 + 1000 * (eig - np.min(eig)) / (np.ptp(eig) + 1e-6)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray',
        node_size=node_sizes, font_size=10, font_weight='bold')

nx.draw_networkx_nodes(G, pos,
                       nodelist=top_indices.tolist(),
                       node_color='orange',
                       node_size=1500,
                       label='Top 5')

plt.title("Eigenvector Centrality (Top 5 Highlighted)")
plt.legend()
plt.show()
```

## Katz Centrality

Katz Centrality extends eigenvector centrality by adding a constant influence factor.

It is defined as:

$$
x_i = \alpha \sum_{j} A_{ij} x_j + \beta
$$

Or in matrix form:

$$
\mathbf{x} = (\mathbf{I} - \alpha A)^{-1} \beta \mathbf{1}
$$

Where:
- $\alpha$ is the attenuation factor (must be less than $1/\lambda_{\text{max}}$,
- $\beta$ is the baseline centrality contribution,
- $A$ is the adjacency matrix.

This allows all nodes to have non-zero centrality, even if they are isolated.


```python
katz = KC(G)

top_indices = np.argsort(katz)[-5:][::-1]
for i in top_indices:
    print(f"Node {i + 1}: Katz Centrality = {katz[i]:.4f}")

pos = nx.spring_layout(G, seed=42)
node_sizes = 300 + 1000 * (katz - np.min(katz)) / (np.ptp(katz) + 1e-6)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray',
        node_size=node_sizes, font_size=10, font_weight='bold')

nx.draw_networkx_nodes(G, pos,
                       nodelist=top_indices.tolist(),
                       node_color='orchid',
                       node_size=1500,
                       label='Top 5')

plt.title("Katz Centrality (Top 5 Highlighted)")
plt.legend()
plt.show()
```

## Random Walk Closeness Centrality (MFPT)

Random Walk Closeness Centrality is based on the concept of **Mean First Passage Time (MFPT)** — the expected number of steps a random walker takes to reach a node from all other nodes.

It is defined as:

$$
C_{RW}(i) = \frac{N - 1}{\sum_{j \neq i} \text{MFPT}_{ji}}
$$

Where:

- $N$ is the total number of nodes in the graph  
- $\text{MFPT}_{ji}$ is the mean first passage time from node $j$ to node $i$

---

### 📐 Matrix Formulation

Let:

- $P$ be the **transition probability matrix**
- $\mathbf{w}$ be the **stationary distribution**, with $\mathbf{W} = \mathbf{w} \mathbf{1}^T$
- $Z$ be the **fundamental matrix**, defined as:

$$
Z = (I - P + W)^{-1}
$$

Then the **MFPT from node $j$ to node $i$** is:

$$
\text{MFPT}_{ji} = \frac{Z_{ii} - Z_{ji}}{w_i}
$$

Where $w_i$ is the stationary probability of node $i$.

This formulation considers all paths — not just the shortest — and incorporates random walk dynamics.


```python
rw_clo = closeness_centrality_mfpt(G)  

top_indices = np.argsort(rw_clo)[-5:][::-1]
for i in top_indices:
    print(f"Node {i + 1}: RW Closeness (MFPT) = {rw_clo[i]:.4f}")

pos = nx.spring_layout(G, seed=42)
node_sizes = 300 + 1000 * (rw_clo - np.min(rw_clo)) / (np.ptp(rw_clo) + 1e-6)

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='gray',
        node_size=node_sizes, font_size=10, font_weight='bold')

nx.draw_networkx_nodes(G, pos,
                       nodelist=top_indices.tolist(),
                       node_color='mediumturquoise',
                       node_size=1500,
                       label='Top 5')

plt.title("Random Walk Closeness (MFPT) Centrality (Top 5 Highlighted)")
plt.legend()
plt.show()
```

## Assortativity and Disassortativity in Networks

**Assortativity** in a network refers to the tendency of nodes to connect to other nodes that are similar to themselves based on some attribute (e.g., degree, clustering coefficient, centrality, etc.).

In contrast, **disassortativity** occurs when nodes tend to connect to others that are different from themselves.

---

###  Mathematical Definition

Assortativity is often quantified using the **Pearson correlation coefficient** between the values of a given node attribute (e.g., degree) and the average of the same attribute across its neighbors.

- If the correlation is **positive**, the network is **assortative** with respect to that attribute.
- If the correlation is **negative**, the network is **disassortative**.
- A correlation near **zero** implies no particular preference (neutral mixing).

---


###  Why It Matters

Understanding assortativity helps reveal:
- **Robustness** and **vulnerability** in networks
- The **spread of information or disease**
- The **structure of social or technological systems**

Social networks tend to be assortative, while biological and technological networks often show disassortativity.


```python
clustering = nx.clustering(G)

avg_neighbor_clustering = {}
for node in G.nodes():
    neighbors = list(G.neighbors(node))
    if neighbors:
        avg = np.mean([clustering[n] for n in neighbors])
    else:
        avg = np.nan
    avg_neighbor_clustering[node] = avg

nodes = list(G.nodes())
x_vals = np.array([clustering[n] for n in nodes])
y_vals = np.array([avg_neighbor_clustering[n] for n in nodes])


plt.figure(figsize=(8, 6))
plt.scatter(x_vals, y_vals, color='royalblue', alpha=0.7)
plt.xlabel("Node Clustering Coefficient")
plt.ylabel("Avg Neighbor Clustering Coefficient")
plt.title("Clustering Assortativity in Karate Club Network")
plt.grid(True)
plt.show()

x_mean = np.mean(x_vals)
y_mean = np.mean(y_vals)
numerator = np.sum((x_vals - x_mean) * (y_vals - y_mean))
denominator = np.sqrt(np.sum((x_vals - x_mean)**2)) * np.sqrt(np.sum((y_vals - y_mean)**2))
r = numerator / denominator

print(f"Pearson correlation: r = {r:.4f}")
```

## Network Percolation

1. **Goal**  
   - Start with a network of $N$ nodes and a set of edges.  
   - Randomly remove parts of the network to see how it fragments.

2. **Two Types of Percolation**  
   - **Bond percolation** $(edges)$:  
     - For each edge, **keep** it with probability $p$.  
     - **Remove** it with probability $1-p$.  
   - **Site percolation** (nodes):  
     - For each node, **keep** it (and all its incident edges) with probability $p$.  
     - **Remove** it (and its incident edges) with probability $1-p$.

3. **Measuring Connectivity**  
   - After removal, find the **largest connected component** $(LCC)$.  
   - Let its size be $L$.

4. **Order Parameter**  
   $$
     S(p) \;=\; \frac{L}{N}
   $$  
   - Fraction of nodes in the LCC.  
   - Tracks how “intact” the network remains as \(p\) varies.

5. **Phase Transition**  
   - **Low $p$**: network is highly fragmented ⇒ $S(p)\approx 0$.  
   - **Critical $p_c$**: a “giant” component suddenly appears ⇒ $S(p)$ jumps up.  
   - **High $p$**: most of the network stays connected ⇒ $S(p)\to 1$.

---


## Percolation Threshold in an Erdős–Rényi graph

Suppose we start with an Erdős–Rényi graph $G(N,p_0)$, where each of the $\tfrac{N(N-1)}{2}$ possible edges is present independently with probability $p_0$.  

1. **Initial average degree**  
   $$
     \langle k_0\rangle \;=\; p_0\,(N-1).
  $$

2. **Bond percolation with occupation probability $p$**  
   After percolation, each existing edge is retained independently with probability $p$.  The new average degree becomes  
   $$
     \langle k(p)\rangle
     \;=\;
     p \,\langle k_0\rangle
     \;=\;
     p\,p_0\,(N-1).
   $$

3. **Condition for a giant component**  
   A giant connected component appears when the mean degree crosses unity:  
   $$
     \langle k(p)\rangle = 1.
   $$

4. **Critical occupation probability**  
   Solving $p\,p_0\,(N-1)=1$ gives  
   $$
     p_c \;=\;\frac{1}{p_0\,(N-1)}.
   $$

   - In the special case $p_0=1$ (i.e  percolation on a complete graph), this reduces to  
     $$
       p_c = \frac{1}{N-1}\approx\frac{1}{N}.
     $$




```python

N  = 200     
p0 = 0.05   

G = nx.erdos_renyi_graph(N, p0)
pos = nx.spring_layout(G)  
```

```python


def bond_percolation(G, p):
    
    H = G.copy()
    for u, v in list(G.edges()):
        if random.random() > p:
            H.remove_edge(u, v)
    return H
```

```python
p = 0.4
H = bond_percolation(G, p)

plt.figure(figsize=(5,5))
nx.draw(G, pos, node_size=20, alpha=0.2, edge_color='gray')
nx.draw(H, pos, node_size=20, edge_color='orange')
plt.title(f"Bond percolation at p = {p:.2f}")
plt.axis('off')
plt.show()
```

```python
ps = np.linspace(0, 1, 21) 
S  = []                    
n_trials = 30             

for p in ps:
    sizes = []
    for _ in range(n_trials):
        H = bond_percolation(G, p)
        if len(H):
            L = max(map(len, nx.connected_components(H)))
        else:
            L = 0
        sizes.append(L / N)
    S.append(np.mean(sizes))

plt.figure(figsize=(6,4))
plt.plot(ps, S, 'o-', label='Empirical $S(p)$')

mean_k0 = p0 * (N - 1)
p_c = 1 / mean_k0
plt.axvline(p_c, color='gray', linestyle='--',
            label=f'$p_c\\approx1/\\langle k\\rangle={p_c:.2f}$')
plt.xlabel('Bond‐occupation probability $p$')
plt.ylabel('Relative LCC size $S(p)$')
plt.legend()
plt.show()

```
